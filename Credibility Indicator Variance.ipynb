```python
def argRelVariance(df):
    argRel = df.loc[:, ['quiz_task_uuid', 'article_number', 'quiz_taskrun_uuid', 
                        'contributor_uuid', 'T1.Q1']]
    argRelGrouped = (argRel.loc[:,  ['quiz_task_uuid', 'article_number', 'T1.Q1']]
                 .groupby(['article_number', 'quiz_task_uuid', 'T1.Q1'])['T1.Q1'].count()
                 .to_frame()
                 .rename({'T1.Q1':'count'}, axis=1))
    varDict = {}
    argRelCount = argRelGrouped.reset_index()
    for num in argRelCount['article_number'].unique():
        varDict[num] = {}
    for i in range(len(argRelCount)):
        varDict[argRelCount.iloc[i, 0]][argRelCount.iloc[i, 1]] = []
    for i in range(len(argRelCount)):
        varDict[argRelCount.iloc[i, 0]][argRelCount.iloc[i, 1]].append(argRelCount.iloc[i, 3])
    for article in varDict.keys():
        for text in varDict[article].keys():
            varDict[article][text] = np.var(varDict[article][text])
    finalDict = {}
    for article in varDict.keys():
        finalDict[article] = []
        for text in varDict[article].keys():
           finalDict[article].append(varDict[article][text])
    for article in finalDict.keys():
        finalDict[article] = np.mean(finalDict[article])
    return finalDict
```
